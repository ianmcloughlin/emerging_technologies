# Single Qubits

***

```python
# Quantum Computing.
import qiskit

# Histograms.
import qiskit.visualization as viz
```

```python
# Quantum Computer Simulator.
simulator = qiskit.Aer.get_backend('aer_simulator')
```

```python
# Design a Quantum Circuit with 1 qubit and 1 classical bit.
qc = qiskit.QuantumCircuit(1, 1)

# Python's output.
qc
```

```python
# Draw circuit using matplotlib (mpl).
qc.draw(output='mpl')
```

```python
# Add a measurement from qubit 0 to classical bit 0.
qc.measure(0, 0)
```

```python
# Draw the circuit with the measurement included.
qc.draw(output='mpl')
```

```python
# Compile the designed circuit for the simulator.
compiled_qc = qiskit.transpile(qc, simulator)
```

```python
# Run the simulated circuit 1024 times by default.
result = simulator.run(compiled_qc).result()

# Display the number of 0's and nubmer of 1's measured (1024 in total).
result.get_counts()
```

## $\textsf{X}$ Gate (Not)

***

```python
# Quantum circuit.
qc = qiskit.QuantumCircuit(1, 1)

# Place X gate on qubit 0.
qc.x(0)

# Incldue a measurement of the qubit.
qc.measure(0, 0)

# Draw the circuit.
qc.draw(output='mpl')
```

```python
# Compile circuit.
compiled_qc = qiskit.transpile(qc, simulator)

# Run simulation.
result = simulator.run(compiled_qc).result()

# Display result.
result.get_counts()
```

## Double $\textsf{X}$ gate

***

```python
# Circuit.
qc = qiskit.QuantumCircuit(1, 1)

# X Gate.
qc.x(0)

# Another X gate.
qc.x(0)

# Measure.
qc.measure(0, 0)

# Draw.
qc.draw(output='mpl')
```

```python
# Compile.
compiled_qc = qiskit.transpile(qc, simulator)

# Simulate.
result = simulator.run(compiled_qc).result()

# Results.
result.get_counts()
```

## $\textsf{H}$ Gate

***

```python
# Circuit.
qc = qiskit.QuantumCircuit(1, 1)

# H Gate.
qc.h(0)

# Measure.
qc.measure(0, 0)

# Draw.
qc.draw(output='mpl')
```

### Matrices

***

$\ket{0} = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \qquad \ket{1} = \begin{bmatrix} 0 \\ 1 \end{bmatrix} $

$\textsf{X} = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$

$\textsf{H} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}$

$\textsf{H} \textsf{X} \ket{0} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix}$

$\textsf{H} \textsf{X} = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix} =  \begin{bmatrix} 1 & 1 \\ -1 & 1 \end{bmatrix}$

$\textsf{H} \textsf{X} \ket{0} = \begin{bmatrix} 1 & 1 \\ -1 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{-1}{\sqrt{2}} \end{bmatrix}$
