```python
import random

import itertools as it
```

```python
[random.choice([0, 1]) for _ in range(8)]
```

```python
def create_f():
  outputs = [random.choice([0, 1]) for _ in range(8)]
  def f(x1, x2, x3):
    i = x1 * 2**2 + x2 * 2**1 + x3 * 2**0
    return outputs[i]
  return f
```

```python
my_f = create_f()
```

```python
my_f(0, 0, 0)
```

```python
my_f(0, 0, 1)
```

```python
list(it.product([0, 1], repeat=3))
```

```python
for inputs in it.product([0, 1], repeat=3):
  print(f"{inputs} -> {my_f(*inputs)}")
```
